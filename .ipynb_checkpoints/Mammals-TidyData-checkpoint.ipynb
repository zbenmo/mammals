```python
from pathlib import Path
import pandas as pd
import re

from scipy.io.arff import loadarff
```

```python
path_with_files = Path(r"C:\Users\zbenm\Mammals")
```

```python
file = "mammals.arff"
#file = "commonmammals.arff"

data, meta = loadarff(path_with_files / file)
```

```python
df = pd.DataFrame(data); df.shape
```

```python
df.head()
```

```python
columns = df.columns

r = re.compile('^bio')
bio_columns = [col for col in columns if r.match(col)]

r = re.compile('^[A-Z]')
mammal_columns = [col for col in columns if r.match(col)]

grid_cell_columns = [col for col in columns if col not in set(bio_columns) | set(mammal_columns)]
```

```python
df[bio_columns].head()
```

```python
df[mammal_columns].head()
```

```python
df['cell_id'] = df.index
```

```python
df_grid_cell_v1 = df[['cell_id'] + grid_cell_columns]; df_grid_cell_v1.shape
```

```python
df_grid_cell_v1.head()
```

```python
df_grid_cell_v2 = df_grid_cell_v1.melt(id_vars=['cell_id', 'latitude', 'longitude']); df_grid_cell_v2.shape
```

```python
df_grid_cell_v2.head()
```

```python
df_grid_cell_v2[['statistics', 'month']] = pd.DataFrame.from_records(
    df_grid_cell_v2['variable'].str.split('_').apply(lambda l: ('_'.join(l[:-2]), '_'.join(l[-2:])))
); df_grid_cell_v2.shape
```

```python
df_grid_cell_v2.head()
```

```python
df_grid_cell_v3 = df_grid_cell_v2.pivot(values='value',
                   index=['cell_id', 'latitude', 'longitude', 'month'],
               columns=['statistics']).reset_index(); df_grid_cell_v3.shape
```
